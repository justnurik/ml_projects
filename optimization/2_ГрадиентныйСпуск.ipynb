__Задача 1. (всего 5 баллов)__ Рассмотрим следуюущую задачу оптимизации:

\begin{equation}
\min_{x \in \mathbb{R}^d} \left[f(x) := \tfrac{1}{2} x^T A x - b^T x\right],
\end{equation}
где $A \in \mathbb{R}^{d \times d}$, $b \in \mathbb{R}^d$.

__а). (0.5 балла)__ Докажите, что константа сильной выпуклости для этой задачи $\mu = \frac{1}{2} \lambda_{\min} \left(A + A^T\right)$, где $\lambda_{\min}(A)$ - минимальное собственное значение матрицы $A$.

Запишем определение $\mu$-сильной выпуклости:
    
- $f(y) \geq f(x) + \langle \nabla f(x), y - x \rangle + \frac{\mu}{2} \|x - y\|^2,$ где $\nabla f(x) = \frac{1}{2}(A + A^{T})x - b$.

Теперь распишем это выражение подробнее:

- $(y^{T}Ay - 2b^{T}y) - (x^{T}Ax - 2b^{T}x) - \langle (A + A^{T})x - 2b, y - x \rangle \geq \mu \|y - x\|^2$

- $y^{T}Ay - x^{T}Ax - \langle (A + A^{T})x, y - x \rangle \geq \mu \|y - x\|^2$

- $y^{T}Ay - x^{T}Ay - x^{T}A^{T}y + x^{T}A^{T}x \geq \mu \|y - x\|^2$

где можно сгруппировать выражения следующим образом:

- $(y - x)^{T}Ay - x^{T}(A^{T}y - A^{T}x) \geq \mu \|y - x\|^2$

- $\langle Ay, y - x \rangle - \langle Ax, y - x \rangle \geq \mu \|y - x\|^2$

- $\langle A(y - x), y - x \rangle \geq \mu \langle y - x, y - x \rangle$

Теперь введём замену $z = y - x$, и неравенство примет вид:

- $\langle z, Az \rangle \geq \mu z^T z$

В силу того что $\frac{1}{2} z^T(A + A^T)z = z^TAz$, это выражение можно переписать так:

- $z^{T}\left(\frac{1}{2}(A + A^{T})\right)z \geq \mu z^{T}z$

Поскольку вектор $z$ произволен, данное неравенство должно выполняться для любого $z$, включая случай, когда $z$ является собственным вектором матрицы $\frac{1}{2}(A + A^{T})$, соответствующим наименьшему собственному значению этой матрицы (которая, к тому же, является симметричной, следовательно, все её собственные значения вещественны). Подставив такой собственный вектор $z$, получаем точную оценку для $\mu$:

- $\mu \leq \frac{1}{2}\lambda_{\text{min}}(A + A^{T})$

Заметим, что при таком значении $\mu = \frac{1}{2}\lambda_{\text{min⁡}}(A+A^T)$, неравенство для $μ$-сильной выпуклости выполняется:

- $f(y) \geq f(x)+ \langle ∇f(x),y−x \rangle + \frac{\mu}{2} \|y−x\|^2$

Следовательно, это значение $μ$ является наибольшим возможным для того, чтобы данное неравенство оставалось верным для всех $x$ и $y$.

Таким образом, мы заключаем, что $\mu = \frac{1}{2}\lambda_{\text{min⁡}}(A+A^T)$ — это максимальная константа сильной выпуклости для данной задачи.

__б). (0.5 балла)__ Докажите, что константа Липшица для данной задачи $L = \frac{1}{2}\lambda_{\max}\left(A + A^T\right)$, где $\lambda_{max}(A)$ - максимально собственное значние матрицы $A$.

Для доказательства того, что константа Липшица $L$ для данной задачи равна $L = \frac{1}{2}\lambda_{\max}(A + A^T)$, запишем определение $L$-гладкости:

- $\|\nabla f(x) - \nabla f(y)\| \leq L \|x - y\| \quad \forall x, y$.

Подставим выражение для градиента $f(x)$ в случае квадратичной функции:

- $\nabla f(x) = \frac{1}{2}(A + A^T)x - b$.

- $\|\nabla f(x) - \nabla f(y)\| = \| \frac{1}{2}(A + A^T)(x - y) \|$.

- $\| \frac{1}{2}(A + A^T)(x - y) \| \leq L \|x - y\|$.

Сделаем замену $z = x - y$

- $\| \frac{1}{2}(A + A^T)z \| \leq L \|z\|$.


Сократим на $\|z\|$ с обеих сторон (в случае $z \neq 0$):

- $\frac{1}{2} \frac{||(A + A^T)z||}{\|z\|} \leq L$.

Будем рассматривать $z$ единичной длины, т.е. $\|z\| = 1$. Тогда неравенство примет вид:

- $\frac{1}{2} \|(A + A^T)z\| \leq L$.

- $\max_{\|z\| = 1} \|(A + A^T)z\| = \|(A + A^T) \|  = \lambda_{\max}(A + A^T)$.

Последний факт был на семинарах.

Следовательно, $L = \frac{1}{2}\lambda_{\max}(A + A^T)$ — это максимальная константа Липшица для данной задачи.

Константа Липшица для данной задачи действительно равна $L = \frac{1}{2} \lambda_{\max}(A + A^T)$

__в). (1 балл)__ Предложите способ генерации матрицы $А$ так, чтобы $f$ была сильно выпуклой, а также можно было задавать и менять $d, \mu, L$. Докажите, что способ корректен.

Заметим, что собственные значения матрицы __инвариантны__ относительно преобразования вида $A \to QAQ^T$, где $Q$ — это ортогональная матрица.

**Способ генерации**:

1. **Генерация матрицы $B$**:

    $ \ \ \  \ \ \ \ \ B = diag(L, a_1, ... , a_{d-2}, \mu) \ $, тут $a_i \in [\mu, L]$. 
    
    Это гарантирует, что матрица $B$ имеет минимальное собственное значение $\mu$ и максимальное собственное значение $L$, а остальные собственные значения находятся в заданном диапазоне.

2. **Генерация ортогональной матрицы $Q$**:

    Существует несколько методов генерации случайных ортогональных матриц, например, мы будем пользоваться QR-разложением

3. **Получение матрицы $A$**:

    $ \ \ \  \ \ \ \ \ A = QBQ^T$.

    Поскольку $B$ является диагональной матрицей с заданными собственными значениями, и мы используем ортогональное преобразование, матрица $A$ также будет иметь те же собственные значения $L,a_1​,…,a_{d−2}​,\mu$.

**Корректность генерации**:
- очевидно если учесть инвариантность собственных чисел относительно преобразования $A \to QAQ^T$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sps
import seaborn as sns

from tqdm.autonotebook import tqdm

sns.set(style='darkgrid')
%matplotlib inline
```

Фиксируем значение `seed`, чтобы обеспечить воспроизводимость экспериментов.

```python
np.random.seed(42)
```

Функция для генерации нужных нам матриц.

```python
def GenerateMatrix(d, mu, L):
    # Шаг 1: Генерация диагональной матрицы B
    # Генерируем случайные значения a_i в диапазоне [mu, L]
    a_values = np.random.uniform(mu, L, d - 2)

    # Формируем диагональную матрицу B
    B = np.diag([L] + list(a_values) + [mu])

    # Шаг 2: Генерация случайной ортогональной матрицы Q
    # Используем QR-разложение для генерации ортогональной матрицы
    H = np.random.randn(d, d)  # Случайная матрица
    Q, _ = np.linalg.qr(H)      # QR-разложение для получения Q
    
    # Шаг 3: Получение матрицы A
    A = Q @ B @ Q.T  # Преобразование A = Q B Q^T
    
    return A
```

__г). (0.5 балла)__  Реализуйте метод градиетного спуска:
\begin{equation}
x_{k+1} = x_k - \gamma_k \nabla f(x_k),
\end{equation}
для квадратичной функции с матрицей, которую сгенерировали выше, и произвольным ненулевым вектором $b$.

Используйте предложенную функцию для реализации алгоритма и допишите недостающие фрагменты. После чего для проверки загрузите функцию в [контест](https://contest.yandex.ru/contest/66540/enter/).

```python
def GradientDescent(grad, criterion, x_0, eps, max_iter, **params):
    """
    grad(x) - функция, которая считает градиент целевой функции;
    criterion(x) - функция, которая считает критерий;
    x_0 - начальная точка;
    eps - точность сходимости (обычно 1e-8);
    max_iter - количество итераций;
    **params - содержит именнованные гиперпараметры метода:
        params['gamma'] - шаг.
    """

    errors = []

    x_k = np.copy(x_0)
    err_x_0 = criterion(x_k)

    errors.append(criterion(x_k) / err_x_0)
    for k in tqdm(range(max_iter)):

        x_k -= grad(x_k) * params["gamma"](k)

        errors.append(criterion(x_k) / err_x_0)
        if errors[-1] < eps:
            break

    return x_k, errors
```

__д). (0.5 балла)__ Для задачи с $d = 50$, $\mu = 1$ и $L = 100$ найдите точное решение $x^*$ - выпишите формулу для решения в явном виде, а затем, реализуйте, например, с помощью пакетов линейной алгебры (есть встроенный в numpy).

Так как функция выпуклая, необходимым и достаточным условием для глобального минимума является равенство градиента нулю в этой точке.

Тогда точное решение можно получить следующим образом:

- $\nabla f(x^{*}) = \frac{1}{2}(A + A^{T})x^{*} + b = 0$

- $x^{*} = -2(A + A^{T})^{-1} \cdot b$

при условии, что матрица $(A+AT)$ обратима. В предложенном выше способе генерации матрицы $A$, и с учетом того, что $L = 100, \ \mu = 1$ эта матрица всегда обратима, так как $\det{A} = L \cdot \mu \cdot a_1 \cdot a_2 \cdot ... \cdot a_{d - 2} \neq 0$.

 $d = 50$, $ \ \mu = 1$ и $L = 100$ .

```python
L = 100
mu = 1
d = 50
```

Генерируем $A$.

```python
b = np.ones(d)
A = GenerateMatrix(d, mu, L)
```

$f(x) := \frac{1}{2} x^T A x - b^T x$.

```python
def f(x):
    return 0.5 * (x.T @ A @ x) -(b.T @ x)
```

Посчитаем реальный ответ.

```python
real_min = np.linalg.solve(A + A.T, -2 * b)

f(real_min)
```

Фиксируем максимальное количество итераций $10^3$, гиперпараметр $\gamma_{k} = \frac{1}{L}$ и начальную точку $x^0$.

```python
max_iter = 1000
step = lambda _: 1 / L
x_0 = np.zeros(d)
```

Градиент функции $f$.


```python
def grad_f(x):
    return 0.5 * (A + A.T) @ x + b
```

Функция ошибки.

```python
def mse(x):
    z = x - real_min
    return np.dot(z, z)
```

Запустим градиентный спуск.

```python
x_k, errors = GradientDescent(
    grad=grad_f, criterion=mse, x_0=x_0, eps=-1, max_iter=max_iter, gamma=step
)
```

- _Теоретическое обоснование_ : Шаг $\gamma = \frac{2}{\mu + L}$ выводится из оптимальных методов первого порядка для гладких и сильно выпуклых функций. Он обеспечивает баланс между быстрым движением по плоским областям и осторожным движением по более крутым областям, сводя к минимуму общее количество итераций.

- _Слишком маленький шаг_ $\gamma$ : Меньшие размеры шага, например, ближе к $0$ приводит к более медленной сходимости, поскольку каждое обновление не продвигается достаточно далеко к минимуму.

- _Слишком большой шаг_ $\gamma$ : Больший размер шага (близкий к $\frac{2}{L}$) может вызвать колебания или даже расхождения, поскольку обновления выходят за пределы минимума.

$\gamma$ от $\frac{2}{L}$ (не включительно) до $\frac{3}{L}$(включительно).

```python
steps = [0.125 * i / L / 2 for i in range(34, 48 + 1)]
```

Запустим градиентный спуск для всех $\gamma$.

Графики стоит строить только для числа итераций менее 100, так как при большем количестве итераций наблюдается сильная расходимость алгоритма.

```python
errors = {
    step: GradientDescent(
        grad=grad_f, criterion=mse, x_0=x_0, eps=-1, max_iter=50, gamma=lambda _: step
    )[1]
    for step in steps
}

# для \gamma = 2 / (mu + L)
last_exp_error = GradientDescent(
    grad=grad_f,
    criterion=mse,
    x_0=x_0,
    eps=-1,
    max_iter=50,
    gamma=lambda _: 2 / (mu + L),
)[1]
```

Визуализируем это на одном графике.

```python
# Получаем цветовую карту для визуализации (plasma для цветовых градиентов)
cmap = plt.get_cmap("plasma")
# Нормализуем индексы для цвета, чтобы градиент распределился по количеству ошибок
norm = plt.Normalize(vmin=0, vmax=len(errors) - 1)

with sns.axes_style("darkgrid"):
    plt.figure(figsize=[10, 6])

    for i, step in enumerate(steps):
        # Метки отображаем только для четных индексов, чтобы не перегружать график
        label = (
            r"спуски с шагом " + r"$\frac{" + f"{(i + 16) * 0.125}" + r"}{L}$"
            if i % 2 == 0
            else None
        )
        plt.plot(errors[step], alpha=0.7, color=cmap(norm(i)), label=label)

    plt.plot(
        last_exp_error,
        alpha=0.7,
        label=r"спуски с шагом " + r"$\frac{2}{\mu + L}$",
        color="red",
        linestyle="--",
    )

    # Установим логарифмическую шкалу по оси Y
    plt.yscale("log")

    plt.xlabel("Число итераций")
    plt.ylabel("Квадратичная ошибка, log scale")
    plt.title(r"Скорость сходимости градиентного спуска при различных шагах $\gamma$")

    plt.grid(True, linewidth=0.5)
    plt.legend()
    plt.show()
```


Градиентный спуск сходится оптимально при значениях $\gamma$ , близких к $\frac{1}{L}$. Если шаг $\frac{2}{L} > 2$, то итерации начинают расходиться, что мы и видим на графике в виде стремительного роста ошибки при больших итерациях. Это видно по графику, все линии кроме $\gamma = \frac{2}{\mu + L}$ резко возрастают — это следствие слишком большого шага $\gamma$, который выходит за пределы допустимого диапазона. 

**Заключение**:

- Оптимальный размер шага $\gamma = \frac{2}{\mu + L}$ приводит к самой быстрой сходимости, поскольку учитывает как гладкость (постоянная Липшица), так и кривизну (сильная выпуклость) функции. Слишком маленькие или слишком большие шаги либо замедляют процесс, либо вызывают нестабильность.

__г). (0.75 балла)__ Шаг Поляка-Шора: $\gamma_k = \tfrac{f(x^k) - f(x^*)}{\alpha \|\nabla f(x^k)\|^2_2}$, где $\alpha \geq 1$ нужно подобрать отдельно. Какое $\alpha$ дало лучший результат?

Шаг Поляка-Шора.

```python
def gammaPoleShore(alpha):
    return lambda x: -(f(x) - f(real_min)) / (alpha * ((grad_f(x)).T @ grad_f(x)))
```

Модернизируем градиентный спуск.

```python
def GradientDescentUPD(grad, criterion, x_0, eps, max_iter, **params):
    """
    grad(x) - функция, которая считает градиент целевой функции;
    criterion(x) - функция, которая считает критерий;
    x_0 - начальная точка;
    eps - точность сходимости (обычно 1e-8);
    max_iter - количество итераций;
    **params - содержит именнованные гиперпараметры метода:
        params['gamma']: функция, которая задаёт шаг градиентного спуска. 
        Это может быть фиксированное значение или функция от текущей точки x_k.
    """

    errors = []

    x_k = np.copy(x_0)
    err_x_0 = criterion(x_k)

    errors.append(criterion(x_k) / err_x_0)
    for k in range(max_iter):

        x_k -= grad(x_k) * params["gamma"](x_k)

        errors.append(criterion(x_k) / err_x_0)
        if errors[-1] < eps:
            break

    return x_k, errors
```

Фиксирем набор $\alpha$.

```python
alphas = [50 + i * 10 for i in range(5)] + [100 + i * 100 for i in range(10)]
```

Запускаем градиентный спуск.

```python
errors_alpha = {
    alpha: GradientDescentUPD(
        grad=grad_f,
        criterion=mse,
        x_0=x_0,
        eps=-1,
        max_iter=max_iter,
        gamma=gammaPoleShore(alpha),
    )[1]
    for alpha in alphas
}

# для \gamma = 2 / (mu + L)
last_exp_error = GradientDescent(
    grad=grad_f,
    criterion=mse,
    x_0=x_0,
    eps=-1,
    max_iter=max_iter,
    gamma=lambda _: 2 / (mu + L),
)[1]
```

```python
# Получаем цветовую карту для визуализации (plasma для цветовых градиентов)
cmap = plt.get_cmap("plasma")
# Нормализуем индексы для цвета, чтобы градиент распределился по количеству ошибок
norm = plt.Normalize(vmin=0, vmax=len(errors_alpha) - 1)

with sns.axes_style("darkgrid"):
    plt.figure(figsize=[10, 6])

    for i, alpha in enumerate(alphas):
        label = r"$\alpha = " + f"{alpha}" + r"$"
        plt.plot(
            errors_alpha[alpha],
            alpha=0.65,
            color=cmap(norm(i)),
            label=label if i % 2 == 0 else None,
        )

    plt.plot(
        last_exp_error,
        alpha=0.7,
        label=r"спуски с шагом " + r"$\frac{2}{\mu + L}$",
        color="red",
        linestyle="--",
    )

    # Установим логарифмическую шкалу по оси Y
    plt.yscale("log")

    plt.xlabel("Число итераций")
    plt.ylabel("Квадратичная ошибка, log scale")
    plt.title(r"Скорость сходимости градиентного спуска при различных шагах $\alpha$")

    plt.grid(True, linewidth=0.5)
    plt.legend()
    plt.show()
```
