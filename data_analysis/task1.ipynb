#### 3. Классификация методом ближайшего соседа

Ранее в контесте вы уже реализовали функцию поиска ближайших соседей. Теперь давайте применим ее для решения нашей текущей задачи. 

Реализуйте функцию, которая для каждого объекта тестового набора данных найдет $k$ ближайших соседей из обучающего набора данных, и в качестве ответа возьмет наиболее часто встречающееся значение целевого признака среди найденных соседей. Обратите внимание, что соседей необходимо искать только среди объектов обучающего набора данных.

````python
def find_nearest_points(A: np.ndarray, B: np.ndarray, k: int) -> np.ndarray:
    distance_matrix = ((A[:, np.newaxis, :] - B) ** 2).sum(axis=2)

    distance_matrix = np.argsort(distance_matrix, axis=0)
    distance_matrix += 1

    return distance_matrix[0:k, :].T
````

````python
def normalize(features):
    """
    Нормализирует по среднему значению в массиве.
    """
    features["Age"] = features["Age"] / features["Age"].mean()
    features["Fare"] = features["Fare"] / features["Fare"].mean()
````

````python
def knn(features, train_features, train_target, k=1):
    """
    Предсказание целевого признака для данных features -- pandas-таблица данных.
    на основе обучающего набора данных train_features, train_target.
    Возвращает pandas.Series с теми же индексами, что и у features.
    Число k -- количество соседей.
    """
    # Нормализация
    normalize(features)
    normalize(train_features)

    # строим матрицу ближайших точек
    matrix_nearest_points = find_nearest_points(
        train_features[["Age", "Fare"]].to_numpy(),
        features[["Age", "Fare"]].to_numpy(),
        k,
    )
    # для удобства
    matrix_nearest_points = pd.DataFrame(matrix_nearest_points)
    matrix_nearest_points.index = features.index

    # меняем индесацию ибо функция find_nearest_points не учитывает индескацию, и работает с чистыми numpy массивами
    train_target.index = np.arange(1, len(train_target) + 1)

    # метод knn
    i = 0
    predicted = np.empty(features.shape[0], dtype=bool)

    for nearest_point in matrix_nearest_points.to_numpy():
        # число выживших соседей
        survived_count = 0

        for index in nearest_point:
            survived_count += train_target[index]

        predicted[i] = 2 * survived_count >= k
        i += 1

    return pd.Series(predicted, features.index)
````

Постройте классификатор методом ближайшего соседа, используя только признаки `Age` и `Fare`.

#### 4. Классификация логическими правилами

На основе выше проведенного исследования предложите два разных правила в виде решающего дерева, пример которого можете посмотреть в презентации с первой лекции (классификация котиков). Никакие статистические методы использовать не требуется, только логику и графики.

В данной задаче достаточно, если первое дерево будет иметь максимальную глубину 2, а второе &mdash; глубину 1, и при этом не является поддеревом первого. *Примечание*: дерево из одного листа имеет глубину 0.

**Дерево 1:**
```
                Билет стоил больше 50
             +-----------------------> Выжил
             ! 
Пассажир ----+
             !  Билет стоил меньше 50
             +-----------------------> Погиб
```
**Дерево 2:**
```
                           1 или 2 класс
                        +----------------> Выжил
                Женщина !
             +----------+
             !          !     3 класс
             !          +----------------> Погиб
             !
Пассажир ----+
             !              Молодой
             !          +-------------> Выжил
             !  Мужчина !
             +----------+
                        !   Немолод
                        +--------------> Погиб
```

Реализуйте полученные деревья по шаблону

````python
def tree1(features):
    """
    Предсказание целевого признака для данных features -- pandas-таблица данных.
    Возвращает pandas.Series с теми же индексами, что и у features.
    """
    ans = pd.Series(np.where(features["Fare"] >= 50, True, False))
    ans.index = features.index
    return ans
````

````python
def tree2(features):
    """
    Предсказание целевого признака для данных features -- pandas-таблица данных.
    Возвращает pandas.Series с теми же индексами, что и у features.
    """
    arr = np.empty(features.shape[0], dtype=bool)

    # Решающее дерево
    arr[:] = np.where(
        ((features["Sex"] == "female") & (features["Pclass"] != 3))
        | (
            (features["Sex"] != "female")
            & (
                (features["AgeGroup"] == "Подросток")
                | (features["AgeGroup"] == "Младенец")
            )
        ),
        True,
        False,
    )

    ans = pd.Series(arr)
    ans.index = features.index
    return ans
````

#### 5. Качество

Вспомним, что у нас имеется тестовая часть выборки. Самое время ее использовать для того, чтобы оценить, насколько хорошими получились деревья. Предложите какой-нибудь критерий качества.

*Замечание*. Не стоит пытаться искать, какие критерии существуют. В данном случае легко предложить адекватный критерий. Стандартные критерии мы разберем позже, но для начала лучше подумать самостоятельно.

**Критерий:**

Находим число "Правильных ответов" и делим на количество вопросов. Тут под "правильными ответами" понимаются ситуации, когда мы угадали судьбу человека, то есть, выжил ли он или нет.

Реализуйте данный критерий по шаблону

````python
def criteria(target, predicted):
    '''
    Вычисление критерия качества для предсказания predicted, 
    если истинные значения -- target.
    Возвращает одно вещественное число.
    '''
    incorrect_answers = (target ^ predicted).sum()
    right_answer = len(target) - incorrect_answers
    return right_answer / len(target)
````

Оцените качество построенных ранее моделей (два дерева, kNN), используя предложенный критерий.

*Примечание.* Полученные значения не влияют на оценку по задание. Оценивается только корректность и обоснованность решения, а также графики и выводы.

Оценка качества для решающих деревьев.

````python
evaluation_for_tree1 = criteria(target_test, tree1(features_test))
evaluation_for_tree2 = criteria(target_test, tree2(features_test))

print(f"Оценка качества первого дерева: {evaluation_for_tree1 :.2%}")
print(f"Оценка качества второго дерева: {evaluation_for_tree2 :.2%}")
````

Посчитайте значение предложенного ранее критерия качества для метода ближайшего соседа для значений $k$ от 1 до 10. Постройте график зависимости критерия качества от $k$. Как вы можете пояснить полученные результаты?

Оценим модель KNN для к = 0, ... , 10

````python
def test_model(test_count):
    evaluation_for_knn = np.array([criteria(target_test, knn(features_test.copy(), features_train.copy(), target_train.copy(), i)) for i in range(1, test_count + 1)])

    #  Для удобства
    evaluation_for_knn = pd.DataFrame(evaluation_for_knn, index=np.arange(1, test_count + 1), columns=["Result"])
    return evaluation_for_knn
````

````python
with sns.axes_style("darkgrid"):
    evaluation_for_knn = test_model(10)
    plt.figure(figsize=(12, 6))

    sns.lineplot(data=evaluation_for_knn, x=evaluation_for_knn.index, y="Result").set(
        xlabel="Значение k",
        ylabel="Точность Модели",
        title="Точность модели в зависимости от K",
    )

    plt.show()
````

Поведение модели при первых 100 значениях k.
